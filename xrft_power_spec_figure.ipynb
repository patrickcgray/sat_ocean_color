```python
km_per_pix = 111 # roughly...
fig, axes = plt.subplots(nrows=2, ncols=4, figsize=(18,7))
fig.set_tight_layout(True)

data_mab_nj_full.chlor_a.plot(ax=axes[0,0], vmax=1.5, vmin=0, cmap='jet')
data_mab_nj.chlor_a.plot(ax=axes[0,1], vmax=0.3, vmin=0, cmap='jet')
im = axes[0,2].pcolormesh(powerspec.freq_lon/km_per_pix, powerspec.freq_lat/km_per_pix, powerspec, norm=colors.LogNorm())
axes[0,3].plot(iso_powerspec.freq_r/km_per_pix, iso_powerspec, label='Chl-a Variance Spectrum')

# # just plotting normal -5/3 line
x = iso_powerspec.freq_r/km_per_pix
y = np.power(iso_powerspec.freq_r/km_per_pix, -5/3)
axes[0,3].plot(x, y, label='-5/3 power law')

cbar = fig.colorbar(im, ax=axes[0,2])
cbar.set_label('variance')
axes[0,3].legend(loc="upper right")
#axes[0,3].set_ylim(1e1, 1e4)
#axes[0,3].set_xlim(0.002, 1e-1)
axes[0,3].set_xscale('log')
axes[0,3].set_yscale('log')
axes[0,2].set_xlabel(r'k [1/km]')
axes[0,2].set_ylabel(r'l [1/km]')
axes[0,3].set_xlabel(r'k$_r$ [1/km]')
axes[0,3].set_ylabel(r'variance')

## begin sst row

data_mab_nj_sst_full.sst.plot(ax=axes[1,0], vmax=30, vmin=10, cmap='coolwarm')
data_mab_nj_sst.sst.plot(ax=axes[1,1], vmin=18, vmax=25, cmap='coolwarm')
im = axes[1,2].pcolormesh(powerspec_sst.freq_lon/km_per_pix, powerspec_sst.freq_lat/km_per_pix, powerspec_sst, norm=colors.LogNorm())
axes[1,3].plot(iso_powerspec_sst.freq_r/km_per_pix, iso_powerspec_sst, label='SST Variance Spectrum')

# # just plotting normal -5/3 line
x = iso_powerspec_sst.freq_r/km_per_pix
y = np.power(iso_powerspec_sst.freq_r/km_per_pix, -5/3)
axes[1,3].plot(x, y, label='-5/3 power law')

cbar = fig.colorbar(im, ax=axes[1,2])
cbar.set_label('variance')
axes[1,3].set_ylim(1e4, 1e7)
axes[1,3].set_xlim(0.002, 1e-1)
axes[1,3].legend(loc="upper right")
axes[1,3].set_xscale('log')
axes[1,3].set_yscale('log')
axes[1,2].set_xlabel(r'k [1/km]')
axes[1,2].set_ylabel(r'l [1/km]')
axes[1,3].set_xlabel(r'k$_r$ [1/km]')
axes[1,3].set_ylabel(r'variance')

#plt.savefig('spectral_analysis', dpi=600, facecolor='w', edgecolor='w',
#        orientation='landscape', papertype=None, format=None,
#        transparent=True, bbox_inches='tight')
```
