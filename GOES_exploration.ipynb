```python
import matplotlib.pyplot as plt

from skimage.measure import label
from skimage import data
from skimage import color
from skimage.morphology import extrema
from skimage import exposure
```

```python
img = np.copy((ds_gs.sea_surface_temperature[0].values* 100).astype(np.int16))
```

```python
import numpy as np
from skimage.morphology import extrema

#We create an image (quadratic function with a maximum in the center and
#4 additional constant maxima.
#The heights of the maxima are: 1, 21, 41, 61, 81

w = 10
x, y = np.mgrid[0:w,0:w]
f = 20 - 0.2*((x - w/2)**2 + (y-w/2)**2)
f[2:4,2:4] = 40; f[2:4,7:9] = 60; f[7:9,2:4] = 80; f[7:9,7:9] = 100
f = f.astype(np.int)

#We can calculate all maxima with a height of at least 40:

maxima = extrema.h_maxima(f, 40)
```

```python
label_h_minima = label(maxima)
overlay_h = color.label2rgb(label_h_minima, f, alpha=0.7, bg_label=0,
                            bg_color=None, colors=[(1, 0, 0)])
```

```python
fig, ax = plt.subplots(1, 3, figsize=(15, 5))

ax[0].imshow(f, cmap='bwr')
ax[0].set_title('Original image')
ax[0].axis('off')
ax[0].set_ylim(ax[0].get_ylim()[::-1])

ax[1].imshow(overlay_h)
ax[1].set_title('Local Maxima')
ax[1].axis('off')
ax[1].set_ylim(ax[1].get_ylim()[::-1])
```

```python
h = 10
h_minima = extrema.h_minima(img, h)
label_h_minima = label(h_minima)
overlay_h = color.label2rgb(label_h_minima, img, alpha=0.7, bg_label=0,
                            bg_color=None, colors=[(1, 0, 0)])
```

```python
# We find all local maxima
local_minima = extrema.local_minima(img)
label_minima = label(local_minima)
overlay = color.label2rgb(label_minima, img, alpha=0.7, bg_label=0,
                          bg_color=None, colors=[(1, 0, 0)])
```
